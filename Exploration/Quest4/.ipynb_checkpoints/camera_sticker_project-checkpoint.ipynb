```python
import os
import cv2
import matplotlib.pyplot as plt
import numpy as np
import dlib
```

```python
os.getcwd()
```

```python
my_image_path = '/aiffel/aiffel/camera_sticker/images/project_image.jpeg'
img_bgr = cv2.imread(my_image_path)
img_show = img_bgr.copy()
plt.imshow(img_bgr)
plt.show()
```

```python
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.show()
```

```python
detector_hog = dlib.get_frontal_face_detector()

img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
dlib_rects = detector_hog(img_rgb, 1)
```

```python
print(dlib_rects)   

for dlib_rect in dlib_rects: # 찾은 얼굴 영역의 좌표
    l = dlib_rect.left() # 왼쪽
    t = dlib_rect.top() # 위쪽
    r = dlib_rect.right() # 오른쪽
    b = dlib_rect.bottom() # 아래쪽

    cv2.rectangle(img_show, (l,t), (r,b), (0,255,0), 2, lineType=cv2.LINE_AA) # 시작점의 좌표와 종료점 좌표로 직각 사각형을 그림

img_show_rgb =  cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB)
plt.imshow(img_show_rgb)
plt.show()
```

```python
detector_hog = dlib.get_frontal_face_detector()

img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
dlib_rects = detector_hog(img_rgb, 1)
```

```python
print(dlib_rects)   

for dlib_rect in dlib_rects: # 찾은 얼굴 영역의 좌표
    l = dlib_rect.left() # 왼쪽
    t = dlib_rect.top() # 위쪽
    r = dlib_rect.right() # 오른쪽
    b = dlib_rect.bottom() # 아래쪽

    cv2.rectangle(img_show, (l,t), (r,b), (0,255,0), 2, lineType=cv2.LINE_AA) # 시작점의 좌표와 종료점 좌표로 직각 사각형을 그림

img_show_rgb =  cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB)
plt.imshow(img_show_rgb)
plt.show()
```

```python
model_path = '/aiffel/aiffel/camera_sticker/models/shape_predictor_68_face_landmarks.dat'
# 사전 훈련된 모델을 저장해둔 주소
```

```python
landmark_predictor = dlib.shape_predictor(model_path)
```

- shape_predictor 함수를 이용해 모델을 불러온다.
- lanmark_predictor는 RGB이미지와 dlib.rectangle를 입력 받고 dlib.full_object_detection을 반환
- dlib.rectangle : 내부 예측 박스, dlib.full_object_detection : 각 구성 요소의 위치와 이미지 상의 객체의 위치

```python
list_landmarks = [] # landmark 위치 저장 list

for dlib_rect in dlib_rects :
    points = landmark_predictor(img_rgb,dlib_rect) # RGB이미지와 box위치(rectangle)를 입력으로 받은 모델
        # points = 각 landmark의 위치와 이미지 상의 객체의 위치
    list_points = list(map(lambda p:(p.x,p.y), points.parts())) # landmark 위치 정보를 (x,y)형태로 list로 저장
    list_landmarks.append(list_points)

print(len(list_landmarks[0])) # 얼굴 n개의 경우 list_landmarks n개의 원소
                            #각 원소는 68개의 landmark위치가 나열된 list 
print(len(list_landmarks))
```

```python
for landmark in list_landmarks:
    for point in landmark:
        cv2.circle(img_show, point, 2, (0, 0, 0), -1)

img_show_rgb = cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB)
plt.imshow(img_show_rgb)
plt.show()
```

cv2.circle(img_show, point, 2, (0,0,0), -1)
- circle : openCV의 원을 그리는 함수
- img_show의 각각의 point에 크기가 2이고 (0,0,0)색으로 내부가 채워진(-1) 원을 그림 = 점
- 마지막 파라미터가 정수 = 선의 두께 의미

```python
for dlib_rect, landmark in zip(dlib_rects, list_landmarks):
    print(landmark[30]) # 코 중앙의 index = 30
    x = landmark[30][0] # 이미지에서 코 부위의 x값
    y = landmark[30][1] # 이미지에서 코 부위의 y값
    
    w = h = dlib_rect.width() # 얼굴 영역의 가로를 차지하는 픽셀 수
    
    print(f'(x,y) : ({x},{y})')
    print(f'(w,h) : ({w},{h})')
```

```python
refined_x = x - w // 2 
refined_y = y - h // 2

print(f'(x,y) : ({refined_x},{refined_y})')
```

```python
sticker_area = img_show[refined_y:refined_y+img_sticker.shape[0], refined_x:refined_x+img_sticker.shape[1]]

img_show[refined_y:refined_y+img_sticker.shape[0], refined_x:refined_x+img_sticker.shape[1]] = \
    np.where(img_sticker==255,sticker_area,img_sticker).astype(np.uint8)
```

## 스티커 적용

```python
sticker_path = '/aiffel/aiffel/camera_sticker/images/cat_nose.png'
for dlib_rect, landmark in zip(dlib_rects, list_landmarks):
                         # 코 끝의 index = 33
    x = landmark[33][0] # 이미지에서 코 부위의 x값
    y = landmark[33][1] # 이미지에서 코 부위의 y값
    
    w = h = dlib_rect.width()
    
    img_sticker = cv2.imread(sticker_path)

    img_sticker = cv2.resize(img_sticker,(w,h))
    
    # 스티커 붙일 위치 조정
    refined_x = x - w // 2
    refined_y = y - h // 2
    
    # 스티커 붙일 영역
    sticker_area = img_bgr[refined_y:refined_y +img_sticker.shape[0], refined_x:refined_x+img_sticker.shape[1]]
    
    img_bgr[refined_y:refined_y +img_sticker.shape[0], refined_x:refined_x+img_sticker.shape[1]] = \
    np.where(img_sticker==255,sticker_area,img_sticker).astype(np.uint8) # 255영역(=흰색부분)은 원래 이미지 영역으로 남기고 나머지 부분 스티커
    
plt.imshow(cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB))
plt.show()
```

### 회전

```python
import math
```

```python
# 약간 기울어진 얼굴에 따른 스티커 수정
sticker_path = '/aiffel/aiffel/camera_sticker/images/cat_nose.png'
for dlib_rect, landmark in zip(dlib_rects, list_landmarks):
                         # 코 끝의 index = 33
    x = landmark[33][0] # 이미지에서 코 부위의 x값
    y = landmark[33][1] # 이미지에서 코 부위의 y값
    
    w = h = dlib_rect.width()
    
    img_sticker = cv2.imread(sticker_path)
    img_sticker = cv2.resize(img_sticker,(w,h)) # 스티커 사이즈 조절
    sticker_x = img_sticker.shape[0]
    sticker_y = img_sticker.shape[1] # 스티커 x,y 좌표
    
    # 스티커 붙일 위치 조정
    refined_x = x - w // 2
    refined_y = y - h // 2
    
    # 스티커 붙일 영역
    sticker_area = img_bgr[refined_y:refined_y +sticker_x, refined_x:refined_x+sticker_y]
    
    # 얼굴의 회전 각도
    height, width, channel = img_sticker.shape
    ang = math. atan(h/w)/math.pi*180
    print('이미지가 기울어진 각도 : ', -round(ang,2))
    
    # 각도에 맞춰 스티커 회전
    matrix = cv2.getRotationMatrix2D((width/2, height/2),-ang,1)
    img_sticker = cv2.warpAffine(img_sticker, matrix, (width,height))
    
    img_bgr[refined_y:refined_y +sticker_x, refined_x:refined_x+sticker_y] = \
    np.where(img_sticker==255,sticker_area,img_sticker).astype(np.uint8) # 255영역(=흰색부분)은 원래 이미지 영역으로 남기고 나머지 부분 스티커
    
plt.imshow(cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB))
plt.show()
```
